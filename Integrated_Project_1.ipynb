cleaning each name to avoid any hyphens, numbers, extra spaces, and anything after ":". 

```python
def strip_subtitle(name):
    return name.split(':')[0].strip()

def remove_numbers(name):
    return re.sub(r'\b\d+\b', '', name).strip()

def clean_name(name):
    name = strip_subtitle(name)
    name = remove_numbers(name)
    name = name.replace(' - ', ' ')
    return name.strip()
```

```python
base_names = [clean_name(name) for name in game_names]
```

Finding dominant franchise names up to "Call" "of" "Duty"

```python
franchise_threshold = 3
franchise_counter = Counter(base_names)
dominant_franchises = [name for name, count in franchise_counter.items() if count >= franchise_threshold]
```

Grouping game names by dominant franchise or sending to other if no match found.

```python
franchise_groups = defaultdict(list)

for original_name in game_names:
    cleaned_name = clean_name(original_name)
    matched = False

    for franchise in dominant_franchises:
        if franchise in cleaned_name:
            franchise_groups[franchise].append(original_name)
            matched = True
            break

    if not matched:
        franchise_groups['Other'].append(original_name)
```

Counting words that show up often in other category.

```python
other_titles = franchise_groups['Other']

other_word_counter = Counter()
for title in other_titles:
    words = set(re.sub(r'\b\d+\b', '', title).split())
    other_word_counter.update(words)
```

Finding common words in other category from count above.

```python
other_word_threshold = 3
dominant_other_words = [word for word, count in other_word_counter.items() if count >= other_word_threshold]
```

Moving other games into new groups based on common words.

```python
remaining_other_titles = []
new_franchises = defaultdict(list)

for title in other_titles:
    matched = False
    for word in dominant_other_words:
        if word in title:
            new_franchises[word].append(title)
            matched = True
            break
    if not matched:
        remaining_other_titles.append(title)
```

Merging new franchise into main groups.

```python
for word, titles in new_franchises.items():
    franchise_groups[word].extend(titles)

franchise_groups['Other'] = remaining_other_titles
```

Filtering for pokemon variations.

```python
final_groups = defaultdict(list)

for franchise, titles in franchise_groups.items():
    for title in titles:
        if 'pokémon' in title or 'pokemon' in title:
            final_groups['pokemon'].append(title)
        else:
            final_groups[franchise].append(title)
```

Going over unlisted games and listing them to a new frnachise name if the word or phrase comes before roman numerals, numbers, or a semicolon.

```python
roman_numeral_pattern = r'\b(i|ii|iii|iv|v|vi|vii|viii|ix|x)\b'

ungrouped_games = final_groups.get('', [])

potential_franchises = []
franchise_lookup = {}

for game in ungrouped_games:
    game_lower = game.lower()
    match = re.search(r'(\s\d+|\s' + roman_numeral_pattern + r'|:)', game_lower)

    if match:
        franchise_name = game_lower[:match.start()].strip()
        potential_franchises.append(franchise_name)
        franchise_lookup[game] = franchise_name
    else:
        franchise_lookup[game] = None
```

```python
franchise_counts = Counter(potential_franchises)

remaining_games = []

for game, franchise_name in franchise_lookup.items():
    if franchise_name and franchise_counts[franchise_name] >= 2:
        if franchise_name not in final_groups:
            final_groups[franchise_name] = []
        final_groups[franchise_name].append(game)
    else:
        remaining_games.append(game)

final_groups[''] = remaining_games
```

Checking to see how many unlisted games are left.

```python
print(f"Total ungrouped games: {len(final_groups.get('', []))}")
```

Grouping similar bigram franchise names together and merging game names under new franchise name.

```python
stopwords = {'the', 'of'}

bigram_counter = Counter()
franchise_bigrams = {}

for franchise in final_groups.keys():
    words = [word for word in franchise.split() if word not in stopwords]
    bigrams = [' '.join(words[i:i+2]) for i in range(len(words)-1)]
    franchise_bigrams[franchise] = bigrams
    bigram_counter.update(bigrams)

bigram_threshold = 2
dominant_bigrams = {bigram for bigram, count in bigram_counter.items() if count >= bigram_threshold}

merged_franchises = defaultdict(list)

for franchise, bigrams in franchise_bigrams.items():
    matched = False
    for bigram in bigrams:
        if bigram in dominant_bigrams:
            merged_franchises[bigram].extend(final_groups[franchise])
            matched = True
            break
    if not matched:
        merged_franchises[franchise].extend(final_groups[franchise])

final_groups = merged_franchises
```

### Testing Hypothesis

- Xbox One vs. PC User Ratings
    - Null Hypothesis: The average user ratings for XOne and PC games are the same.
    - Alternative Hypothesis: The average user ratings for XOne and PC games are different.

- Action vs. Sports Genre User Ratings
    - Null Hypothesis: The average user ratings for Action and Sports games are the same.
    - Alternative Hypothesis: The average user ratings for Action and Sports games are different.

```python
alpha = 0.05
```

```python
xbox_one_ratings = relevant_data[relevant_data['platform'] == 'XOne']['user_score']
pc_ratings = relevant_data[relevant_data['platform'] == 'PC']['user_score']

action_ratings = relevant_data[relevant_data['genre'] == 'Action']['user_score']
sports_ratings = relevant_data[relevant_data['genre'] == 'Sports']['user_score']
```

```python
platform_test = st.stats.ttest_ind(xbox_one_ratings, pc_ratings, equal_var = False)
genre_test = st.stats.ttest_ind(action_ratings, sports_ratings, equal_var = False)

print(f'Platform test p-value: {platform_test.pvalue}')
print(f'Genre test p-value: {genre_test.pvalue}')

if platform_test.pvalue < alpha:
    print('We reject the null hypothesis, average user ratings are not the same for Xbox One and PC')
else:
    print('We accept the null hypothesis, average user ratings are the same for Xbox One and PC')
    
if genre_test.pvalue < alpha:
    print('We reject the null hypothesis, average user ratings are not the same for action and sports genre')
else:
    print('We accept the null hypothesis, average user ratings are same for action and sports genre')
```

- Alpha: I set the significance level at 0.05 to allow a 5% chance of error, which is standard for hypothesis testing.
- Hypothesis: I compared the means of user ratings between two platforms and two genres, testing if they are statistically the same or different using independent t-tests.
- Why: Understanding differences in ratings helps identify if certain platforms or genres consistently receive better or worse feedback from players.
